# Data Cleaning in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\User\Python_Learning\Pandas\Datasets\Customer Call List.xlsx")
df
```

```python
df = df.drop_duplicates()
df

# the df = df....(whatever changes) will overwrite the data frame with the latest changes
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
#df["Last_Name"] = df["Last_Name"].str.lstrip("...")
#df["Last_Name"] = df["Last_Name"].str.lstrip("/")
#df["Last_Name"] = df["Last_Name"].str.rstrip("_")
#df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df["Last_Name"] = df["Last_Name"].str.replace('[^a-zA-Z0-9]','',regex=True)
df

# strip() by default without any arguments will remove extra spaces only
# we can strip individual items or we can combine them all into one single strip e.g. strip("123._/") means remove any numericals . _ and /
# alternatively we can just replace all symbols with blanks
```

```python
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
#df["Phone_Number"].str.replace('[^a-zA-Z0-9]','',regex=True)
# replace('[^a-zA-Z0-9]','') means replace anything that is not alphabets a-z A-Z or numericals 0-9 with blanks i.e. symbols

#df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
# this will give an error because the phone number needs to be converted into a string first

df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))
df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df
# replace symbols with blank > convert phone numbers to strings > adds '-' into the phone numbers after range 0:3 and after range 3:6 
# > replace nan-- and na-- with blanks
```

```python
df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
#df["Address"].str.split(',',n=2, expand=True)
df

# split (',' , n=2 , expand = True) means split by the comma ',' expected up to 2 commas and expand = True means split into additional columns
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')

df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')
df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')

df
```

```python
df = df.replace('N/a','')
df = df.replace('NaN','')

df=df.fillna('')

df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)
        
df
```

```python
for x in df.index:
    if df.loc[x, "Phone_Number"] == '':
        df.drop(x, inplace=True)

df

#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
df = df.reset_index(drop=True)
df

# resets the indexing, should be done at the end since the cleaning involved dropping data rows
```
